```python
import os
os.getcwd()

import jieba
import pandas as pd
import re
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv('Result.csv')
df
```

```python
STOPWORDS.update(['China', 'Chinese'])
STOPWORDS
```

```python
top_words_15 = []
for t in df['Transcript']:
    words = re.split('[ ,.]', t)
    words = list(filter(lambda x : x != '', words))
    newWords = list(filter(lambda x:x not in STOPWORDS, words))
    
    counts = {}
    ciYun = []
    for word in newWords:
        # == 1 单个词语不计算在内
        if len(word) < 2:
            continue
        else:
            # 遍历所有词语，每出现一次其对应的值加 1
            counts[word] = counts.get(word, 0) + 1
            ciYun.append(word)
            
    # 将键值对转换成列表
    items = list(counts.items())


    # 根据词语出现的次数进行从大到小排序
    items.sort(key=lambda x: x[1], reverse=True)
    top_words_15.append(items[:15])
df['top_words_15'] = top_words_15
df
```

```python
df.to_csv('Result.csv', index=0)
```
