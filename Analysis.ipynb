```python
import os
os.getcwd()

import pandas as pd
from textblob import TextBlob

import numpy as np
```

```python
df1 = pd.read_csv('Result_China.csv')
print(len(df1))
df2 = pd.read_csv('Result_US.csv')
print(len(df2))
```

```python
clean_df1_t = list(filter(lambda x: isinstance(x, str), df1['Transcript']))
print(len(clean_df1_t))
clean_df2_t = list(filter(lambda x: isinstance(x, str), df2['Transcript']))
print(len(clean_df2_t))
```

```python
# df1 统计
data = clean_df1_t
polarity_list = list(map(lambda x: TextBlob(x).sentiment.polarity, data))
subjectivity_list = list(map(lambda x: TextBlob(x).sentiment.subjectivity, data))
                         
max_p = np.max(polarity_list)
print('Max_p: {}'.format(max_p))
min_p = np.min(polarity_list)
print('Min_p: {}'.format(min_p))
mean_p = np.mean(polarity_list)
print('Mean_p: {}'.format(mean_p))
median_p = np.median(polarity_list)
print('Median_p: {}'.format(median_p))
var_p = np.var(polarity_list)
print('Var_p: {}'.format(var_p))

max_s = np.max(subjectivity_list)
print('Max_s: {}'.format(max_s))
min_s = np.min(subjectivity_list)
print('Min_s: {}'.format(min_s))
mean_s = np.mean(subjectivity_list)
print('Mean_s: {}'.format(mean_s))
median_s = np.median(subjectivity_list)
print('Median_s: {}'.format(median_s))
var_s = np.var(subjectivity_list)
print('Var_s: {}'.format(var_s))
```
